```python
    
t5["Scores"]=0

for index,rows in t5.iterrows():
    score=0

for index,rows in t5.iterrows():
    score=0
    if(str(rows[7]).lower().find("b.tech") !=-1 or str(rows[7]).lower().find("b.e") !=-1):
        if(rows[9]==2020):
            score=score+10
        elif(rows[9]==2019):
            score=score+8
        else:
            score=score+5
    elif(str(rows[7]).lower().find("m.sc") !=-1 or str(rows[7]).lower().find("m.tech") !=-1):
        if(rows[9]==2020):
            score=score+7
        else:
            score=score+7
    if(rows[3]==3):
        score=score+10
    elif(rows[3]==2):
        score=score+7
    elif(rows[3]==1):
        score=score+3
    
    if(rows[4]==3):
            score=score+10
    elif(rows[4]==2):
        score=score+7
    elif(rows[4]==1):
        score=score+3
        
    if(rows[2]==3):
        score=score+10
    elif(rows[2]==2):
        score=score+7
    elif(rows[2]==1):
        score=score+3   
        
    print(rows[0])    
    if(str(rows[5]).lower().find('machine learning') !=-1):
        
        score=score+3
    if(str(rows[5]).lower().find('deep learning') !=-1):
        score=score+3  
    if(str(rows[5]).lower().find('nlp') !=-1 or str(rows[5]).lower().find('natural language processing') !=-1) :
        score=score+3
    if(str(rows[5]).lower().find('aws') !=-1 or str(rows[5]).lower().find('amazon web services') !=-1) :
        score=score+3 
    if(str(rows[5]).lower().find('excel') !=-1):
        score=score+3
    t5.iloc[index,-1] =score 
    
```

```python
# Split a data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Decison tree model
from sklearn.tree import DecisionTreeClassifier
dt_classifier=DecisionTreeClassifier()
```

```python
dt_classifier.fit(X_train,y_train)
dt_classifier.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix 
```

```python
y_pred1 = dt_classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred1)
```

```python
finalreport = classification_report(y_test, y_pred1,output_dict=True)
```

```python
def predict_score(Application_ID,Degree,Year_Graduation,Scores,City,Python,R,DataScience,otherskill):    


    X = np.zeros(9)
    X[0] = Scores
    X[1] = Application_ID
    X[2] = Degree
    X[3] = Year_Graduation
    X[4] = City
    X[5] = Python
    X[6] = R
    X[7] = DataScience
    X[8] = otherskill
    return model.predict([X])[0]
```
